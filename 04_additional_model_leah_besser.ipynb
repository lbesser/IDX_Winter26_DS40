```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

# Load Data

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
y_train = train['ClosePrice']
y_test = test['ClosePrice']
```

```python
# removed identity-based categorical encodings to reduce overfitting and improve generalization

cols_to_drop = [col for col in train.columns
                if col.startswith('CoListOfficeName_')
                or col.startswith('BuyerOfficeName_')
                or col.startswith('BuyerAgentAOR_')
                or col.startswith('ListAgentFullName_')
                or col.startswith('ListAgentAOR_')
                or col.startswith('Flooring_')]

X_train = train.drop(columns=cols_to_drop)
X_test = test.drop(columns=cols_to_drop)

X_train = X_train.drop(columns=['OriginalListPrice', 'ListingId','ClosePrice','CloseDate'], errors='ignore')
X_test = X_test.drop(columns=['OriginalListPrice', 'ListingId','ClosePrice','CloseDate'], errors='ignore')
```

```python
print(X_train.select_dtypes(exclude='number').columns)
```

## Impute

```python
imputer = SimpleImputer(strategy='mean')

X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```



# Decision Tree With List Price



```python
tree = DecisionTreeRegressor(random_state=42)

tree.fit(X_train_imputed, y_train)

y_pred_tree = tree.predict(X_test_imputed)

r2_tree = r2_score(y_test, y_pred_tree)
mae_tree = mean_absolute_error(y_test, y_pred_tree)

print("Decision Tree R²:", round(r2_tree, 4))
print("Decision Tree MAE:", round(mae_tree, 2))
```

**Compared to Baseline Model**

Linear Regression R²: 0.976

Linear Regression MAE: $53,031.9
